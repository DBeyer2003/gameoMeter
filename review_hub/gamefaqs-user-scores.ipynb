```python
# open the file for reading one line at a time
filename = "/Users/DBeye/new_django_game/review_csvs/user-date-folder/sonic-secret-rings-dated.csv"

# open the file for reading
f = open(filename,encoding="utf8") 
# discard the first line containing headers
headers = f.readline()

#started variables.
game_id = 25103
platform = "Wii"
current_start_date = "2007-02-20"
#stores the initial numbers of ratings as variables, so that we can keep track of
#how many we already have for future dates.
initial_rating_count = {
  10:0,9:0,8:0,7:0,6:0,
  5:0,4:0,3:0,2:0,1:0
}
overall_rating_count = 0

#used to store each batch of user ratings.
user_ratings = {}

# go through the entire file one line at a time
for line in f:

  try:
    fields = line.split(',')
    #print(start_date)
    current_end_date = fields[0]

    rating_info = {  
             10:int(fields[1]),
             9:int(fields[2]),8:int(fields[3]),
             7:int(fields[4]),6:int(fields[5]),
             5:int(fields[6]),4:int(fields[7]),
             3:int(fields[8]), 2:int(fields[9]),
             1:int(fields[10])}
    
    #print(current_end_date,rating_info[10],rating_info[9],rating_info[8],rating_info[7],rating_info[6],rating_info[5],rating_info[4],rating_info[3],rating_info[2],rating_info[1])
    rating_list = list(rating_info.keys())

    #converted dates.
    conv_start_date = datetime.strptime(current_start_date, '%Y-%m-%d')
    conv_end_date = datetime.strptime(current_end_date, '%Y-%m-%d')
    #measurements used to create a rate-of-publication, used to determine the rate
    #at which reviews are published.
    days_bt_dates = (conv_end_date-conv_start_date).days
    #for each rating, we will add to the initial values, and use that those to subtract
    #from the rating infos in each new line. That way, we only interate on each batch
    #of ratings within the date range, excluding ratings that are already added before.
    total_ratings = 0
    for t in range(10):
      total_ratings += (rating_info[t+1]-initial_rating_count[t+1])
      rating_info[t+1] = rating_info[t+1]-initial_rating_count[t+1]
      #print("For srart date ", conv_start_date , "and end-date ", conv_end_date , "By taking ", rating_info[t+1], " and subtracting ", initial_rating_count[t+1], " to get ", rating_info[t+1]-initial_rating_count[t+1], " we currently have total ratings of ", total_ratings)
    rate_of_pub = 1.0/float(total_ratings/days_bt_dates)


    

      #print(list(rating_permutations))

    
    #iterates through all of the ratings.
    for current_r in range(total_ratings):
      day_to_publish = math.ceil(float(current_r+1)*(rate_of_pub))
      rand_day = 0 
      rand_date = fields[0]
      #we will select a date within the specified amount of days before and/or after
      #the date. (E.g. if we're publishing on day 4, and the rate
      # of publication is 2 we'll choose a day in the range from
      #day 2 to day 6.) If day_to_publish is day 1, we'll just publish on this day.
      if day_to_publish > 1:
        #choose a random date within the given range.
        rand_day = random.randint(day_to_publish-(math.ceil(rate_of_pub)-1),day_to_publish+(math.ceil(rate_of_pub)-1))
        rand_date = conv_start_date+timedelta(rand_day)
      else: 
        rand_day = random.randint(1,day_to_publish+(math.ceil(rate_of_pub)-1))
        rand_date = conv_start_date+timedelta(rand_day)
      
      #choose a random rating to modify; only choose this rating if
      # there are still ratings that we need to add for this number.
      rand_rating = random.choice(rating_list)
      #ensure that we don't go under 0 by choosing a different number if necessary.
      
      #add to the initial numbers to increment.

      #used to generate fake user name, with length from 5 to 20.
      length = 20
      random_user = ''.join(random.choices(string.ascii_letters + string.digits, k=random.randint(5,length)))
      
      #stores the score/user info in the dictionary. Because we're iterating across
      #different lines we'll use the overall rating count as a key, rather than
      #the number that we're iterating towards.
      user_ratings[overall_rating_count] = {'rating': rand_rating,'date_published':str(rand_date.strftime('%Y-%m-%d')),
                        'user':random_user,'is_user':False, 
                        'platform':platform,'game_id':game_id}
      #print(user_ratings)
      
      #remove the current rating from the rating_dict if there is nothing else to add.
      #and increment initial values so that we can subtract them from later values.
      rating_info[rand_rating] -= 1
      #increment the rating count for the current star rating
      # AND the overall rating count.
      initial_rating_count[rand_rating] += 1
      overall_rating_count += 1
      #delete the rating from the current dict/list if there is no more to add.
      if rating_info[rand_rating] <= 0:
        del rating_info[rand_rating]
        rating_list.remove(rand_rating)
    
      #at the end, we'll set the start_date to the current one for reference
      #with later dates.
      #print(user_ratings)
    current_start_date = current_end_date
  except:
    print("WAAAAAHIIEEEEYYYYYYYYOOOOOO-OOOO")






```
