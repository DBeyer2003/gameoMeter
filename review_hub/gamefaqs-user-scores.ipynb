```python
#get url response
import urllib
import json
import time
from datetime import *
import math
import random
import string

#saves reviews in dataframe
import pandas as pd
```

```python
reference_info = {
                  'game-id':25103,
                  'platform':"Wii",
                  'start-date':"2007-02-20",
                  'end-date':"2019-01-26",
}

rating_info = {  
             10:40,
             9:58,8:155,
             7:254,6:233,
             5:130,4:103,
             3:62, 2:51,
             1:35}
rating_list = list(rating_info.keys())

conv_start_date = datetime.strptime(reference_info['start-date'], '%Y-%m-%d')
conv_end_date = datetime.strptime(reference_info['end-date'], '%Y-%m-%d')
days_bt_dates = (conv_end_date-conv_start_date).days
total_ratings = rating_info[10]+rating_info[9]+rating_info[8]+rating_info[7]+rating_info[6]+rating_info[5]+rating_info[4]+rating_info[3]+rating_info[2]+rating_info[1]
rate_of_pub = 1.0/float(total_ratings/days_bt_dates)

if (rate_of_pub) % 1 >= 0.5:
  print(days_bt_dates,total_ratings,math.ceil(rate_of_pub), conv_start_date, conv_end_date)
else:
  print(days_bt_dates,total_ratings,round(rate_of_pub), conv_start_date, conv_end_date)

user_ratings = {}

for current_r in range(total_ratings):
  day_to_publish = math.ceil(float(current_r+1)*(rate_of_pub))
  rand_day = 0 
  rand_date = reference_info['start-date']
  #we will select a date within the specified amount of days before and/or after
  #the date. (E.g. if we're publishing on day 4, and the rate
  # of publication is 2 we'll choose a day in the range from
  #day 2 to day 6.) If day_to_publish is day 1, we'll just publish on this day.
  if day_to_publish > 1:
    #choose a random date within the
    rand_day = random.randint(day_to_publish-(math.ceil(rate_of_pub)-1),day_to_publish+(math.ceil(rate_of_pub)-1))
    rand_date = conv_start_date+timedelta(rand_day)

  else: 
    rand_day = random.randint(1,day_to_publish+(math.ceil(rate_of_pub)-1))
    rand_date = conv_start_date+timedelta(rand_day)
  
  #choose a random rating to modify; only choose this rating if
  # there are still ratings that we need to add for this number.
  rand_rating = random.choice(rating_list)

  #used to generate fake user name, with length from 5 to 20.
  length = 20
  random_user = ''.join(random.choices(string.ascii_letters + string.digits, k=random.randint(5,length)))
  
  #stores the score/user info in the dictionary.
  user_ratings[current_r] = {'rating': rand_rating,'date_published':str(rand_date.strftime('%Y-%m-%d')),
                     'user':random_user,'is_user':False, 
                     'platform':reference_info['platform'],'game_id':reference_info['game-id']}
  rating_info[rand_rating] -= 1


  #remove the current rating from the rating_dict if there is nothing else to add.
  if rating_info[rand_rating] <= 0:
    del rating_info[rand_rating]
    rating_list.remove(rand_rating)

print(user_ratings)
  

  
  
  
  
```

```python
import csv 

csv_filename = "sonic-secret-rings-gamefaqs.csv"

fieldnames = ['rating','date_published','user','is_user','platform','game_id']

with open(csv_filename,mode='w',newline='') as file:
  writer = csv.DictWriter(file,fieldnames=fieldnames)
  writer.writeheader()
  for r in user_ratings:
    writer.writerow({field: user_ratings[r].get(field) or r for field in fieldnames})
```
